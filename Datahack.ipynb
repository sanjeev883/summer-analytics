```python
from sklearn.preprocessing import OrdinalEncoder

# Define categories for each column
age_categories = ['18 - 34 Years', '35 - 44 Years', '45 - 54 Years', '55 - 64 Years', '65+ Years']
education_categories = ['< 12 Years', '12 Years', 'Some college', 'College graduate']
income_categories = ['Below Poverty', '<= $75,000, Above Poverty', '> $75,000']

# Create OrdinalEncoder objects for each column
enc_age = OrdinalEncoder(categories=[age_categories])
enc_education = OrdinalEncoder(categories=[education_categories])
enc_income = OrdinalEncoder(categories=[income_categories])

# Fit and transform all columns in one go
df['age_group'] = enc_age.fit_transform(df[['age_group']])
df['education'] =enc_education.fit_transform(df[['education']])
df['income_poverty'] = enc_income.fit_transform(df[['income_poverty']])
```
